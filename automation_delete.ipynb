```python
from pydrive2.auth import GoogleAuth
from pydrive2.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials
import datetime
import time
```

```python
auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive = GoogleDrive(gauth)
```

```python
checkpoints_folder_id = '14ngYm_C52_yDVoC3UERQC9mpGklYbCzS'
file_list1 = drive.ListFile({'q': f"'{checkpoints_folder_id}' in parents and trashed=false"}).GetList()
for file1 in file_list1:
  print(file1['title'])
while True:
  file_to_deletes = []
  check1 = False
  file_list1 = drive.ListFile({'q': f"'{checkpoints_folder_id}' in parents and trashed=false"}).GetList()
  for file1 in file_list1:
    if str(file1['title']).find('.pkl') != -1:
      # print(str(file1['title']))
      check1 = True

  file_list2 = drive.ListFile({'q': f"'{checkpoints_folder_id}' in parents and trashed=true"}).GetList()
  for file1 in file_list2:
    if str(file1['title']).find('.pkl') != -1:
      # print(str(file1['title']))
      file_to_deletes.append(file1)

  if check1 and len(file_to_deletes) > 0:
    for file_to_delete in file_to_deletes:
      file_to_delete.Delete()
      title = file_to_delete['title']
      now = datetime.datetime.now()
      formatted = now.strftime("%d-%m-%Y %H:%M:%S")

      print(f'Delete file {title} at {formatted}')
  time.sleep(10*60)
```
